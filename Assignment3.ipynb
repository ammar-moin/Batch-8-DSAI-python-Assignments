A- Discounts and Pricing
1. Check Discount Eligibility
Write a program to check if a customer is eligible for a discount. If the
total purchase is
more than $100, apply a 10% discount and display the final price.
Otherwise, display the
total price as it is.

```python
discount = 0.10
total_purchase  = float(input("Enter the total purchase: $ "))
if total_purchase > 100:
  discount_price = total_purchase * discount
  final_price = total_purchase - discount_price
  print(f"A 10% discount has been applied. FInal price: ${final_price:.2f}")
else:
  print(f"No discount applied. Total Price: $ {total_purchase:.2f}")

```

2. Calculate Bulk Discount
If a customer buys more than 5 items, apply a 15% discount on the
total price. Otherwise,
no discount is applied. Display the total price.

```python
discount_rate = 0.15
item_purchased = int(input("Enter the numer of items purchased: "))
total_price = float(input("Enter the total price of items: $ "))

if item_purchased > 5:
  final_price = total_price - (total_price * discount_rate)
  print(f"The Discount of 15% has benn applied. Final price: {final_price:.2f}")
else:
  print(f"No discount applied. Total Price: {total_price:.2f}")
```

3. Membership Discount
Check if the customer is a member (is_member = True). Members get
a 20% discount;
non-members get a 5% discount. Calculate and print the discounted
price.

```python
total_price = float(input("Enter the total price: $ "))
is_member_input =  input("Does the customer have a membership? (yes/no): ").strip().lower()

is_member = True if is_member_input == 'yes' else False

if is_member:
  final_price = total_price - (total_price* 0.20)
  print(f"The discount of 5% is applied for Members Final price: $ {final_price:.2f}")
else:
  final_price = total_price - (total_price* 0.05)
  print(f"The discount of 5% is applied for non-members. Final price: $ {final_price:.2f}")
```

16. Travel Discount
If a person is traveling more than 500 miles, offer a 20% discount on
ticket price.
Otherwise, charge the full amount.

```python
distance = float(input("Enter the distance you're travelling (in miles): "))
ticket_price = float(input("Enter the ticket price: $ "))

if distance > 500:
  final_price = ticket_price - (ticket_price*0.20)
  print("Hurray! You have got a discount of 20% on your ticket.")
else:
  final_price = ticket_price

print(f"The final price for ticket is: ${final_price:.2f}")
```
