```python
# Find and display the largest number of a list without using built-in function max(). Your program
# should ask the user to input values in list from keyboard.

user_input = input("Enter the elements of the list separated by spaces: ")
input_list = [int(num) for num in user_input.split()]
largest = input_list[0]
for num in input_list:
  if num > largest:
    largest = num

print(f"The largest number in list is: {largest}")
```

8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this :

```python
m = int(input("Enter the number of rows (m): "))
n = int(input("Enter the number of columns (n): "))

matrix = []
print("Enter the elements of the matrix row by row: ")
for i in range(m):
  row = list(map(int,input(f"Enter row {i+1} (space-sperated): ").split()))
  if len(row) != n:
    print(f"Error: please enter exactly, {n} elements for this row")
    exit()
  matrix.append(row)


for i in range(m):
  row_sum = sum(matrix[i])
  print(f"Sum of row {i+1} = {row_sum}")

```

9. Write a program to add two matrices of size n x m.

```python
m = int(input("Enter the number of rows (m): "))
n = int(input("Enter the number of columns (n): "))

matrix1 = []
matrix2 = []

print("Enter the elements of the first matrix row by row: ")
for i in range(m):
  row = list(map(int,input(f"Enter row {i+1} of first matrix (space- separated): ").split()))
  if len(row) != n:
    print(f"Error: Please enter exactly, {n} elements fot this row")
    exit()
  matrix1.append(row)


print("Enter the elements of the second matrix row by row: ")
for i in range(m):
  row = list(map(int,input(f"Enter row {i+1} of second matrix (space- separated): ").split()))
  if len(row) != n:
    print(f"Error: Please enter exactly, {n} elements fot this row")
    exit()
  matrix2.append(row)


result_matrix = []

for i in range(m):
  row_sum = []
  for j in range(n):
    row_sum.append(matrix1[i][j]+ matrix2[i][j])
  result_matrix.append(row_sum)


print("\n the sum of the two matrices is: ")
for row in result_matrix:
  print(" ".join(map(str,row)))
```

10. Write a program to multiply two matrices

```python
m = int(input("Enter the number of rows (m): "))
n = int(input("Enter the number of columns (n): "))

matrix1 = []
matrix2 = []

print("Enter the elements of the first matrix row by row: ")
for i in range(m):
  row = list(map(int,input(f"Enter row {i+1} of first matrix (space- separated): ").split()))
  if len(row) != n:
    print(f"Error: Please enter exactly, {n} elements fot this row")
    exit()
  matrix1.append(row)


print("Enter the elements of the second matrix row by row: ")
for i in range(m):
  row = list(map(int,input(f"Enter row {i+1} of second matrix (space- separated): ").split()))
  if len(row) != n:
    print(f"Error: Please enter exactly, {n} elements fot this row")
    exit()
  matrix2.append(row)


result_matrix = []

for i in range(m):
  row_prod = []
  for j in range(n):
    row_prod.append(matrix1[i][j]*matrix2[i][j])
  result_matrix.append(row_prod)

print("\n The product of two matrices is: ")
for row in result_matrix:
  print(" ".join(map(str,row)))
```
